```python
import pandas as pd
```

```python
# read UCLA Law Covid-19 Behind Bars Data Project in data from https://github.com/uclalawcovid19behindbars/data
UCLA_df = pd.read_csv('ConfirmedCasesandDeaths_061020_UCLALawCovid19BehindBarsDataProject.csv')
```

```python
# read State and Federal Adult Correctional Facilities from https://www.icpsr.umich.edu/web/NACJD/studies/37294
prison_df = pd.read_csv('ICPSRPRISONDATA.csv')
```

```python
# function to parse facility name for join
def facility_simplify(row):
    if type(row) == list:
        return (row[0])
    else:
        return row
```

```python
# prepare UCLA dataframe for join
UCLA_df['Name_join'] = UCLA_df['Name'].str.split(" ")
UCLA_df['Name_join'] = UCLA_df['Name_join'].apply(facility_simplify)
UCLA_df['Name_join'] = UCLA_df['Name_join'].str.upper()
UCLA_df['State'] = UCLA_df['State'].str.upper()
```

```python
# prepare ICPST dataframe for join
prison_df['Name_join'] = prison_df['Name'].str.split(" ")
prison_df['Name_join'] = prison_df['Name_join'].apply(facility_simplify)
prison_df['Name_join'] = prison_df['Name_join'].str.upper()
```

```python
# cleanup columns for join
prison_df = prison_df.rename(columns={'state':'State'})
prison_df['State'] = prison_df['State'].str.upper()
prison_df = prison_df.drop_duplicates(subset='Name_join')
```

```python
# perform join
prison_geo = pd.merge(prison_df, UCLA_df, on=['Name_join', 'State'], how='right')
```

```python
# drop columns used in join
prison_geo = prison_geo.drop(columns=['Name_x', 'Name_join', 'Unnamed: 6'])
```
